## 3. Статистичний аналіз досвіду респондентів

```python
# Пропущені значення були виключені з аналізу, оскільки mean(), median(), mode() у Pandas за замовчуванням ігнорують NaN

mean_exp = data['WorkExp'].mean()
median_exp = data['WorkExp'].median()
mode_exp = data['WorkExp'].mode()[0]

stats_table = pd.DataFrame({
    'Metric': ['Mean', 'Median', 'Mode'], 
    'WorkExp': [round(mean_exp, 2), median_exp, mode_exp]})

print("Work Experience statistics:")
print(stats_table)
```

```python
# розрахунок відсотка популярності Python від респондентів, які відповіли на питання про мови програмування (не від всіх респондентів)
python_users = data['LanguageHaveWorkedWith'].str.contains('Python', na=False).sum()

total_respondents = data['LanguageHaveWorkedWith'].notna().sum()

python_percent = python_users / total_respondents * 100

print(f"Python users: {python_users}")
print(f"Percentage of Python users:  {python_percent:.1f}%")
```

## 7. Географічний аналіз компенсації Python-розробників

```python
[col for col in data.columns
if 'country' in col.lower()]
```

```python
python_dev = data[data['LanguageHaveWorkedWith'].str.contains('Python', na=False)].copy()

python_dev['ConvertedCompYearly'] = pd.to_numeric(python_dev['ConvertedCompYearly'], errors='coerce')

python_dev_clean = python_dev.dropna(subset = ['ConvertedCompYearly'])

comp_by_country = (python_dev_clean.groupby('Country')['ConvertedCompYearly']
                  .agg(mean_compensation='mean', median_compensation='median'))

comp_by_country['mean_compensation'] = comp_by_country['mean_compensation'].round(0)

comp_by_country
```

```python
comp_by_country.sort_values(by='mean_compensation', ascending=False).head(15)
```

## 9. Аналіз популярності Python по віковим категоріям 

```python
python_dev_by_age = (python_dev.groupby('Age')['Age'].agg('count'))
dev_by_age = (data.groupby('Age')['Age'].agg('count'))
```

```python
age_stats = pd.DataFrame({'python_count': python_dev_by_age, 'total_count': dev_by_age})

age_stats['python_%'] = (age_stats['python_count'] / age_stats['total_count'] * 100).round(1)

age_stats
```
